```python
tourism_df.dtypes
```

```python
from pprint import pprint

#total_country_list is our list of country dicts
total_country_list = []

'''
current_country_entry is the per-country dictionary that we build
(and eventually append) to our total list of dicts that we described above.
'''
current_country_entry = {}
doing_inbound  = False
doing_outbound = False

for index, row in tourism_df.iterrows():
    #check to see if we have a country name
    if 'country_name' in current_country_entry.keys():
        #check to see if we are already doing inbound tourism stuff
        if doing_inbound:
            #check to see if this is a row we care about
            if row['Description'] == 'Arrivals - Thousands':
                if row['Series'] == 'TF':
                    current_country_entry['arrivals']['tf'] = list(tourism_df.loc[index, '1995':'2014'])
                elif row['Series'] == 'VF':
                    current_country_entry['arrivals']['vf'] = list(tourism_df.loc[index, '1995':'2014'])
                elif row['Series'] == 'THS':
                    current_country_entry['arrivals']['ths'] = list(tourism_df.loc[index, '1995':'2014'])
                elif row['Series'] == 'TCE':
                    current_country_entry['arrivals']['tce'] = list(tourism_df.loc[index, '1995':'2014'])
                elif row['Series'] == '..':
                    current_country_entry['arrivals']['..'] = list(tourism_df.loc[index, '1995':'2014'])
            elif row['Description'] == 'Tourism expenditure in the country - US$ Mn':
                current_country_entry['expenditure_in_country'] = list(tourism_df.loc[index, '1995':'2014'])
            elif row['Description'] == 'Outbound tourism':
                current_country_entry['arrivals']['total'] = [sum(x) for x in zip(*[current_country_entry['arrivals'][y] for y in current_country_entry['arrivals'].keys()])]
                current_country_entry['departures'] = {} #i changed 'outbound' to 'departures'
                doing_inbound = False
                doing_outbound = True
            continue
        elif doing_outbound:
            #check to see if this is a row we care about
            if row['Description'] == 'Departures - Thousands':
                if row['Series'] == 'TF':
                    current_country_entry['departures']['tf'] = list(tourism_df.loc[index, '1995':'2014'])
                elif row['Series'] == 'VF':
                    current_country_entry['departures']['vf'] = list(tourism_df.loc[index, '1995':'2014'])
                elif row['Series'] == '..':
                    current_country_entry['departures']['vf'] = list(tourism_df.loc[index, '1995':'2014'])
            elif row['Description'] == 'Tourism expenditure in other countries - US$ Mn':
                current_country_entry['expenditure_out_country'] = list(tourism_df.loc[index, '1995':'2014'])
            elif row['Description'] == 'Passenger transport - US$ Mn':
                current_country_entry['departures']['total'] = [sum(x) for x in zip(*[current_country_entry['departures'][y] for y in current_country_entry['departures'].keys()])]
                
                #TODO: current_country_entry = fill_in_empties(current_country_entry)
                current_country_entry['years'] =['1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014']
                total_country_list.append(current_country_entry)
                print('-----------------------------------------')
                print(current_country_entry)
                print('-----------------------------------------')
                current_country_entry = {}
                doing_inbound = False
                doing_outbound = False
        #check to see if we should *start* inbound
        elif row['Description'] == 'Inbound tourism':
            doing_inbound = True
            doing_outbound = False
            current_country_entry['arrivals'] = {}
            continue
    else:
        if str(row['Description']).isupper():
            current_country_entry['country_name'] = str(row['Description'])
```

```python
import json
with open('country_info.json', 'w') as outfile:
    json.dump(total_country_list, outfile, indent =4)
```
